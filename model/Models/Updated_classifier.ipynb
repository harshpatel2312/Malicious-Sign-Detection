```python
import os
import skimage
import pickle
from skimage.io import imread
from skimage.transform import resize
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
input_dir = 'D:/Capstone/carla_lights/traffic_light_data'
categories = ['green', 'red', 'yellow'] 
```

```python
data_train = []
labels_train = []
data_val = []
labels_val = []
```

```python
#Loading training data

for category in categories:
    category_path = os.path.join(input_dir, 'train', category)  # Load from train directory
    for file in os.listdir(category_path):
        img_path = os.path.join(category_path, file)
        img = imread(img_path)
        img = resize(img, (15, 15))  # Resize images to 15x15
        if img.shape == (15, 15, 3):  # Check if the image has the correct shape
            data_train.append(img.flatten())
            labels_train.append(category)  # Use the folder name as the label
        else:
            print(f"Skipping image {file} due to incorrect shape: {img.shape}")
```

```python
#Loading validation data

for category in categories:
    category_path = os.path.join(input_dir, 'val', category)  # Load from val directory
    for file in os.listdir(category_path):
        img_path = os.path.join(category_path, file)
        img = imread(img_path)
        img = resize(img, (15, 15))  # Resize images to 15x15
        print(f"Original shape: {img.shape}")  # Debugging line
        if img.shape == (15, 15, 3):  # Check if the image has the correct shape
            data_val.append(img.flatten())
            labels_val.append(category)  # Use the folder name as the label
        else:
            print(f"Skipping image {file} due to incorrect shape: {img.shape}")
```

```python
x_train = np.asarray(data_train)
y_train = np.asarray(labels_train)
x_val = np.asarray(data_val)
y_val = np.asarray(labels_val)
```

```python
classifier = SVC()

parameters = [{'gamma': [0.01, 0.001, 0.0001], 'C': [1, 10, 100, 1000]}]

grid_search = GridSearchCV(classifier, parameters)

grid_search.fit(x_train, y_train)
```

```python

best_estimator = grid_search.best_estimator_

y_prediction = best_estimator.predict(x_val)

score = accuracy_score(y_prediction, y_val)

print('{}% of samples were correctly classified'.format(str(score * 100)))
```
